## 來進階一點
算個回歸直線如何?

```python
fig, ax = plt.subplots()
rng = np.random.RandomState(1)
x =  rng.rand(100)*3
y =  3 * x - 5 + rng.randn(100)
ax.scatter(x, y);
plt.show()
```

$mx+b = y$

我們把所有的點帶進去，會得到一連串的方程式

排起來之後會變成$Aw=z$

```python
A = np.c_[x.reshape(len(x), 1), np.ones((len(x),1))]
```

```python
z = y.reshape(len(y),1)
```

求解的話

$A^TAw = A^Tz$

```python
w = np.matmul(np.linalg.inv(np.matmul(A.T, A)), np.matmul(A.T, z))
```

```python
fig, ax = plt.subplots()
rng = np.random.RandomState(1)
ax.scatter(x, y);
ax.plot(x, w[0]*x+w[1],'r')
plt.show()
```
